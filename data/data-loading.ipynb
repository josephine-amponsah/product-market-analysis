```python
import json
import pandas as pd
import numpy as np
```

```python
new_cols = ["appDescription", "information", "purchase"]
```

```python
def get_info(file_path, columns):
    with open(file_path, encoding="utf-8") as json_file:
        json_text = json.load(json_file)
        apps_data = {}
        for item in json_text:
            app_data = {}
            for col in columns:
                if col in item.keys() and col == "information":
                    check = "inAppPurchases"
                    if check in item["information"].keys():
                        app_data[check] = item[col]["inAppPurchases"]
                    else:
                        app_data[check] = None
                elif col in item.keys():
                    app_data[col] = item[col]
                else:
                    app_data[col] = None
            apps_data[item["id"]] = app_data  # Assuming "app_name" is a unique identifier for each app
    return apps_data
```

```python
app_info = get_info("app_store_apps.json", new_cols)
```

```python
app_cols = ["appId", "title", "primaryGenre","contentRating", "released", "price", "currency", "developer", "developerId", "score", "reviews", ""]
```

```python
def get_apps(file_path, columns):
    with open(file_path, encoding="utf-8") as json_file:
        json_text = json.load(json_file)
        apps_data = {}
        for item in json_text:
            app_data = {}
            for col in columns:
                if col in item.keys():
                    app_data[col] = item[col]
                else:
                    app_data[col] = None
            apps_data[item["id"]] = app_data  # Assuming "app_name" is a unique identifier for each app
    return apps_data
```

```python
apple_apps = get_apps("apple-apps.json", app_cols)
```

```python
apps = pd.DataFrame.from_dict(apple_apps, orient= "index")
apps = apps.reset_index()
apps= apps.rename(columns={"index":"id"})
```
